# Advanced Multimodal Practice with Google Gemini
Testing Vision-to-Recipe extraction and ingredient recognition.

```python
import google.generativeai as genai
import PIL.Image
import os
from dotenv import load_dotenv

load_dotenv()
genai.configure(api_key=os.getenv("GOOGLE_API_KEY"))
model = genai.GenerativeModel('gemini-1.5-flash')
```

## 1. Ingredient Recognition from Photo
Identify items on a kitchen counter.

```python
def identify_ingredients(image_path):
    img = PIL.Image.open(image_path)
    response = model.generate_content(["List all edible ingredients you see in this photo. Be specific.", img])
    return response.text

# sample = identify_ingredients("countertop.jpg")
print("Gemini Vision Output: [Placeholder for actual image test]")
```

## 2. Recipe Extraction from Cookbook Image
Convert a photo of a page into structured JSON.

```python
def extract_structured_recipe(image_path):
    img = PIL.Image.open(image_path)
    prompt = """
    Extract the recipe from this image into JSON format with the following keys:
    - title
    - ingredients (list of objects with name, amount, unit)
    - steps (list of instructions)
    - total_time_minutes
    """
    response = model.generate_content([prompt, img])
    return response.text

print("Gemini JSON Extraction: [Placeholder for actual image test]")
```

## 3. Visual Cooking Guidance
Ask Gemini if a steak looks done or if onions are caramelized.

```python
def check_cooking_status(image_path, goal):
    img = PIL.Image.open(image_path)
    response = model.generate_content([f"Based on this image, {goal}? Explain why.", img])
    return response.text

# check_cooking_status("pan.jpg", "are these onions caramelized enough for French Onion Soup?")
print("Gemini Chef Consultation: [Placeholder]")
```

## 4. Visual Timer Inference
Infer how much time is left based on the state of the food.

```python
def infer_timer_seconds(image_path, dish_type):
    img = PIL.Image.open(image_path)
    prompt = f"This is {dish_type}. Based on its current appearance in the image, how many more seconds should it cook? Return only a number."
    response = model.generate_content([prompt, img])
    return response.text

print("Gemini Timer Inference: [Placeholder]")
```
